```python
import PIL
from PIL import Image
import numpy as np
import cv2 as cv
import os
from pathlib import Path
```

```python
acess_path = "\\saggital"
temporarysave_path = r"C:\Users\793282\Desktop\MR-Net Data\temporary\train\sag"
save_path = r"C:\Users\793282\Desktop\MR-Net Data\final\train\sag"
```

```python
for file in os.listdir(acess_path):
    #convert to jpg   
    image = np.load(file)
    image = image[0,:,:]
    data = Image.fromarray(image)
    
    name_path = Path(acess_path + '\\' + file).stem
    name = name_path + '.jpg'
    
    data = data.save(temporarysave_path + '\\' + name)
```

```python
for file in enumerate(os.list(temporarysave_path)):
    image = cv.imread(temporarysave_path + '\\' + name)
    LP = cv.Laplacian(image,cv.CV_64F)
    cv.imwrite(save_path + '\\' + file,LP)
```
